```python
import librosa as lb
import tensorflow as tf
import tensorflow.keras.layers as tfkl
from keras import layers, models, losses

from scipy.signal import stft, istft

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

# clases custom

```python
class Spectrogram(tfkl.Layer):
    def __init__(self,win_size,hop_size,fft_size=None,calculate='magnitude',window=tf.signal.hann_window,pad_end=False,name=None, trainable=False):
        super(Spectrogram, self).__init__(name=name)

        self.stft_args = {'ws': win_size,
                  'hs': hop_size,
                  'ffts': fft_size,
                  'win': window,
                  'pad': pad_end,
                  'calculate': calculate}

    def call(self,x):
        stft = tf.signal.stft(
                signals=x,
                frame_length=self.stft_args['ws'],
                frame_step=self.stft_args['hs'],
                fft_length=self.stft_args['ffts'],
                window_fn=self.stft_args['win'],
                pad_end=self.stft_args['pad'])

        calculate = self.stft_args['calculate']
        if calculate == 'magnitude':
            return tf.abs(stft)
        elif calculate == 'complex':
            return stft
        elif calculate == 'phase':
            return tf.math.angle(stft)
        else:
            raise Exception("{} not recognized as calculate parameter".format(calculate))

    def get_config(self):
        config = super().get_config().copy()
        config.update({
            'win_size': self.stft_args['ws'],
            'hop_size': self.stft_args['hs'],
            'fft_size': self.stft_args['ffts'],
            'calculate': self.stft_args['calculate'],
            'window': self.stft_args['win'],
            'pad_end': self.stft_args['pad']
        })
        return config

class SoftMask(tfkl.Layer):
    def __init__(self,name=None):
        super(SoftMask,self).__init__(name=name)

    def call(self,x):
        sources = x[0]
        to_mask = x[1]
        total_sources = tf.expand_dims(tf.reduce_sum(sources,axis=1),axis=1)
        mask = sources/(total_sources+1e-9)
        to_mask = tf.expand_dims(to_mask,axis=1)
        return mask*to_mask

class TranslateRange(tfkl.Layer):
    def __init__(self,name=None,trainable=False,original_range=None,target_range=None):
        super(TranslateRange,self).__init__(name=name)
        self.original_range = original_range
        self.target_range = target_range

    def call(self,x):
        offset = self.target_range[0] - self.original_range[0]
        x_range = self.original_range[1] - self.original_range[0]
        y_range = self.target_range[1] - self.target_range[0]

        return y_range*(x + offset)/x_range

class MSE(tfkl.Layer):
    #x[0]: predicted
    #x[1]: original
    def __init__(self,name=None,trainable=False,lnorm=2,offset=1e-9,normalize=False):
        super(MSE,self).__init__(name=name)
        self.offset = offset
        self.normalize = normalize
        self.lnorm = lnorm
    
    def call(self,x):
        mse_error = tf.abs(x[0] - x[1])**self.lnorm
        if self.normalize:
            mse_error = mse_error/(self.offset + tf.abs(x[1])**self.lnorm)
        return mse_error
```

# Generacion de datos

Genero senos a frecuencias random. Armo un set limpio, y un set con las mismas dimensiones pero le sumo ruido aleatorio. Esta bueno variar el valor de noisy_coef para hacer sets mas o menos ruidosos y ver que pasa. Para que no me estalle la cafetera y probar cosas rapido, armo sets de 128 ejemplos. Esto se puede estirar hasta donde te de la RAM. 

```python
def dummy_seno(duracion, muestras, freq, noisy=False, noisy_coef=0.5):
    t = np.linspace(0, duracion, muestras)
    seno = 0.5 * np.sin(2*np.pi*freq*t)
    if noisy:
        seno = seno + np.random.uniform(low = -noisy_coef, high = noisy_coef, size = seno.shape)
    return seno

freqs = np.random.uniform(low=80, high=8000, size=(1024))
#freqs = 440 * np.ones(128) #para probar cosas 

train_clean = np.array([dummy_seno(duracion = 2, muestras = 32000, freq = f) for i,f in enumerate(freqs)])
train_noisy = np.array([dummy_seno(duracion = 2, muestras = 32000, freq = f, noisy = True) for i,f in enumerate(freqs)])
```

# Armando un modelo funcional (Tensorflow)

La idea es armar un autoencoder basico. Para eso, la red consiste puramente de capas convoucionales. La parte del encoder reduce las dimensiones de entrada aplicando capas convolucionales con strides mayores a 1. La parte del decoder aplica el proceso inverso. 

La red va a tener como entrada el audio limpio y el audio ruidoso. Con el audio ruidoso, calcula la STFT y se queda solo con la magnitud, lo pasa a escala logaritmica y transforma el rango para que quede en valores cercanos al intervalo [0,1] (normalizar por el maximo es una opcion mas simple pero no siempre mas efectiva). Luego arreglo las dimensiones la STFT para que me queden numeros pares y no tener quilombo de dimensiones a lo largo de la red. Para ser precisio, descarto el ultimo frame de tiempo y el ultimo bin de frecuencia (la frecuencia mas alta, no me afecta mucho). Luego se mete esta 'imagen' al encoder,y la salida del encoder se mete al decoder. La salida del decoder se multiplica punto a punto al espectro de entrada. Esta parte se puede entender como la aplicacion de la mascara de amplitud. Lo que resulta de esa magnitud es lo que se va a comparar con la magnitud STFT del audio limpio. Esta comparacion se da en la capa custom 'MSE' que basicamente calcula el error cuadratico medio entre las dos imagenes resultantes (el espectro mascareado, y el espectro limpio) (Deje de aclarar que es la magnitud del espectro y no el espectro completo porque me hinche los huevos). La salida de la red es el resultao de este error. El error se calcula punto a punto en la imagen, osea que tendra una 'imagen de error' (esta bueno para poder visualizarla despues y ver donde esta pifeando, en lugar de ver solo un numero como en la loss que es mas dificil interpretarlo). Esta salida es lo que yo quiero minimizar, pero eso lo veremos en el siguiente bloque 

```python
eps = np.finfo(float).eps #precision de punto flotante
```

```python
x = tfkl.Input((32000), name = 'Entrada_Noisy')
y = tfkl.Input((32000), name = 'Target_Clean')

#GENERACION DE ESPECTROS
#phase_x = Spectrogram(1024,512, calculate = 'phase', name='PHASE_X')(x)
spec_x = Spectrogram(1024,512,name='STFT_X')(x)
spec_x = tf.math.log(spec_x+eps)
spec_x = TranslateRange(original_range=[-5, 5],target_range=[0,1.0])(spec_x)
spec_x = tf.expand_dims(spec_x,axis=-1, name= 'expand_X')
spec_x = tfkl.Cropping2D(((0,1),(0,1)), name = 'Crop_X')(spec_x)

spec_y = Spectrogram(1024,512,name='STFT_Y')(y)
#spec_y = tf.math.log(spec_y+eps) 
#spec_y = TranslateRange(original_range=[-5,5],target_range=[0,1.0])(spec_y)
spec_y = tf.expand_dims(spec_y,axis=-1, name='expand_Y')
#spec_y = tfkl.Cropping2D(((0,1),(0,0)), name='Crop_Y')(spec_y)


#ENCODER
enc = tfkl.Conv2D(16, kernel_size=(3,3), padding='SAME', activation='relu', input_shape=(64,64,1), name='NET1')(spec_x)
#enc = tfkl.Conv2D(16, kernel_size=(3,3), padding='SAME', activation='relu', name='NET2')(enc)####
enc = tfkl.Conv2D(16, kernel_size=(3,3), strides=2, padding='SAME', activation='relu', name='NET3')(enc)
#enc = tfkl.Conv2D(32, kernel_size=(3,3), padding='SAME', activation='relu', name='NET4')(enc)####
enc = tfkl.Conv2D(32, kernel_size=(3,3), strides=2, padding='SAME', activation='relu', name='NET5')(enc)
#enc = tfkl.Conv2D(32, kernel_size=(3,3), padding='SAME', activation='relu', name='NET6')(enc)####

#DECODER
dec = tfkl.Conv2DTranspose(32, kernel_size=(3,3), strides=2, padding='SAME', activation='relu', name='NET7')(enc)
#dec = tfkl.Conv2D(32, kernel_size=(3,3), padding='SAME', activation='relu', name='NET8')(dec)###
dec = tfkl.Conv2DTranspose(16, kernel_size=(3,3), strides=2, padding='SAME', activation='relu', name='NET9')(dec)
#dec = tfkl.Conv2D(16, kernel_size=(3,3), padding='SAME', activation='relu', name='NET10')(dec)###
dec = tfkl.Conv2D(1, kernel_size=(3,3), padding='SAME', activation='relu', name='NET11')(dec)

estimated = tfkl.Multiply()([dec,spec_x])
estimated = tf.pad(estimated,((0,0),(0,1),(0,1),(0,0)), name='SALIDA')


err = MSE()([estimated,spec_y])

stft_calc = tf.keras.Model(inputs=[x,y],outputs=[err])

#Las lineas comentadas forman una red mas compleja, es para probar como varia el rindimiento y demas
```

Falta resolver una cuestion. Como la salida del sistema que arme es enrealidad un frame de loss, en lugar de pasarle un tensor target a la red para que compare con la salida, lo que voy a hacer es mandarle cualquier cosa a la salida (porque no la voy a usar) y voy a armar una funcion de loss propia en donde simplemente se calcule el valor medio de toda ese frame de loss que tengo como salida de la red (paso de un frame a un escalar). Esto puede sonar medio enroscado, pero es una solucion copada por varioas cuestiones: para poder visualizar la loss, para poder trabajar solo con audios y no tener que estar calculando stfts afuera, etc. De todos modos esto se puede cambiar sin problemas. Por ejemplo, puedo salir de la red con el resultado de la multiplicacion entre la mascara y el espectro de entrada. Si hago eso, a la hora de entrenar la red voy a tener que pasarle como salida los frames de STFT de la seÃ±al limpia y ahi podria usar una funcion de loss de las que ya estan predefinidas. Probablemente esa sea la eleccion que se ajuste mejor al proyecyo de denoising, pero como esto es un ejemplo y es mio yo hago lo que quiero asique vamos con la forma 'enroscada'. 

```python
#defino una loss custom
def mean_loss(y_true, y_pred):
  return tf.reduce_mean(y_pred)
```

```python
#defino el optimizador y le indico que use mi loss 
stft_calc.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),loss=mean_loss)
```

# Explorando el modelo y sus resultados

Para ver que esta pasando a lo largo de toda la red, voy a armar esta funcion que basicamente la voy a usar a modo de 'model.predict()', y me va a tirar en una lista todas las activaciones de todas las capas que tenga junto con sus nombres, asi puedo visualizar lo que esta pasando en cada capa. Esta muy bueno. 

```python
def predict(data):
  output = [layer.name for layer in stft_calc.layers]
  outputs = []
  output_names = []
  inputs = []
  input_names = []
  for layer in stft_calc.layers:
      if hasattr(layer,'is_placeholder'):
          inputs.append(layer.output)
          input_names.append(layer.name)
      elif layer.name in output:
          outputs.append(layer.output)
          output_names.append(layer.name)
      else:
          pass
  predict_fn = tf.keras.backend.function(inputs = inputs,outputs=outputs)
  activations = predict_fn(data)
  activations = {name: act for name, act in zip(output_names,activations)}
  return activations
```

# Recuperando Audio

Todo muy lindo, pero ahora hay que ver como escuchamos los resultados del denoiseo furioso y mentiroso. Para eso voy a agarrar la salida de la ultima capa que labura con el espectrograma que aprende la red antes de compararlo con el espectrograma del audio limpio. Este espectro va a ser la amgnitud STFT de mi audio denoiseado. Para poder pasar a audio me falta la info de fase. Aca podemos discutir una vida, pero para hacerla corta vamos a probar con dos cosas: probar usando solo este espectro de magnitud y estimando la fase con el metodo de griffin lim, o usar este espectro de magnitud estimado y la fase del audio sucio para recuperar el audio por grifin lim (siempre griffin lim porque si quiero formar el audio directamente aplicando la ISTFT me puedo encontrar conque forme un espectro incosistente y no puedo pasar al dominio temporal, lo que en cristiano seria que tengo valores negativos de fft. puedo meter una relu ahi o hacer cosas cabezas pero esa es un discusion para otro notebook). 

```python
estim_phase = Spectrogram(1024,512,calculate = 'phase', name='PHASE_X')(train_noisy[0:1]).numpy().reshape(61,513)
estim_magspec = activations['tf_op_layer_SALIDA'].reshape(61,513)

estim_spec = estim_magspec * np.exp(1.0j*estim_phase) #euler s2
```

```python

#para ver como hacer para que estime la fase desde random o desde una fase propuesta lean la docu de esta funcion
estim_audio =  lb.griffinlim(estim_magspec.T, n_iter=1000, #transponer!
                                  hop_length=512, 
                                  win_length=1024, 
                                  window='hann', center=True, 
                                  length=None, 
                                  pad_mode='reflect', 
                                  momentum=0.0, 
                                  init='random', 
                                  random_state=None)
```
